```python
class Metric:
    def __init__(self, name, params, calls, size_before, size_after, exec_before, exec_after, comp_before, comp_after):
        self.name = name
        self.params = params
        self.calls = calls
        self.size_before = size_before
        self.size_after = size_after
        self.size_saving = size_before - size_after
        self.exec_before = exec_before
        self.exec_after = exec_after
        self.exec_saving = exec_before - exec_after
        self.comp_before = comp_before
        self.comp_after = comp_after
        self.comp_saving = comp_before - comp_after
        self.net_saving = self.exec_saving + self.comp_saving - self.comp_before
```

```python
def optimize_specialization(metrics):
    report = {}
    keep = []
    old_gain = 0
    for idx, name in candidates.items():
        if name.startswith('$__imported_wasi_snapshot_preview1_'):
            # print(f'[ignoring] {name}')
            continue

        params = metrics[idx]['params']
        calls = metrics[idx]['calls']
        size_before = metrics[idx]['size_before']
        size_after = metrics[idx]['size_after']
        exec_before = before_exec[idx]
        exec_after = after_exec[idx]
        comp_before = before_comps[idx]
        comp_after = after_comps[idx]
        metric = Metric(name=name,
                        params=params,
                        calls=calls,
                        size_before=size_before,
                        size_after=size_after,
                        exec_before=exec_before,
                        exec_after=exec_after,
                        comp_before=comp_before,
                        comp_after=comp_after)
        report[idx] = metric
        old_gain += metric.exec_saving - metric.comp_after

        if metric.net_saving > 0:
            print(f'[keep] estimated gain {(metric.exec_saving - metric.comp_after) / 1_000_000:3.0f} ms | {name}')
            keep.append(idx)
    gain_sum = sum([report[idx].exec_saving - report[idx].comp_after for idx in keep])
    print(f'#keep {len(keep)} | gain {gain_sum / 1_000_000:.0f} ms | old gain {old_gain / 1_000_000:.0f} ms')
    return report, keep
```

```python
timelines_dir = Path('./out/timelines')
timelines_dir.mkdir(parents=True, exist_ok=True)


def surplus_maximizer(profile):
    waiting_times = profile.virtual_machine.times[:]  # [f, start, end]
    functions = []
    start_time = profile.start
    waiting_times = [(f, s - start_time, e - start_time) for f, s, e in waiting_times]
    surplus = 0
    current_time = 0
    for i, (f, s, e) in enumerate(waiting_times):
        idle = s - current_time
        surplus += idle
        if surplus < e - s:
            functions.append(f)
            surplus += e - s
        current_time = e
        # print(f'{f:4} {s:>10} {e:>10} {surplus:>10}')
    return functions


def plot_timelines(benchmark, profiles):
    max_duration = 0  # (for x-axis)
    max_compilations = 0  # (for y-axis)
    for profile in profiles:
        max_duration = max(max_duration, profile.end - profile.start)
        max_compilations = max(max_compilations, len(profile.compilation_service.times))
    max_x, max_y = max_duration, max_compilations

    fig, axs = plt.subplots(figsize=(10, 1 * len(profiles)), dpi=320,
                            nrows=len(profiles), ncols=1,
                            sharex=True, sharey=True)
    for ax, profile in zip(axs, profiles):
        start = profile.start

        static_start = profile.virtual_machine.static_start
        static_start -= start
        ax.axvline(x=static_start, color='black', linestyle='dashed', linewidth=.5)

        static_end = profile.virtual_machine.static_end
        static_end -= start
        ax.axvline(x=static_end, color='red', linestyle='dashed', linewidth=.5)

        exec_start = profile.virtual_machine.start
        exec_start -= start
        ax.axvline(x=exec_start, color='purple', linestyle='dashed', linewidth=.5)

        exec_end = profile.virtual_machine.end
        exec_end -= start
        ax.axvline(x=exec_end, color='pink', linestyle='dashed', linewidth=.5)

        end = profile.end
        end -= start
        ax.axvline(x=end, color='green', linestyle='dashed', linewidth=.5)

        wasm = profile.wasm_service.times  # [app, start, end]
        wasm = [(c[1], c[2]) for c in wasm]  # [start, end]
        wasm = sorted(wasm, key=lambda x: x[0])  # sort by start
        wasm = [(c[0], c[1] - c[0]) for c in wasm]  # [start, duration]
        wasm = [(c[0] - start, c[1]) for c in wasm]  # subtract start
        for s, d in wasm:
            ax.plot([s, s + d], [0, 0], c='orange', linewidth=1)

        compilations = profile.compilation_service.times  # [app, fn, size, start, end]
        compilations = [(c[2], c[3]) for c in compilations]  # [start, end]
        compilations = sorted(compilations, key=lambda x: x[0])  # sort by start
        compilations = [(c[0], c[1] - c[0]) for c in compilations]  # [start, duration]
        compilations = [(c[0] - start, c[1]) for c in compilations]  # subtract start
        y_offset = .03 * max_compilations
        for y, (s, d) in enumerate(compilations):
            ax.plot([s, s + d], [y + y_offset, y + y_offset], c='blue', linewidth=1)

        latencies = profile.virtual_machine.times  # [f, start, end]
        latencies = [(c[1], c[2]) for c in latencies]  # [start, end]
        latencies = sorted(latencies, key=lambda x: x[0])  # sort by start
        latencies = [(c[0], c[1] - c[0]) for c in latencies]  # [start, duration]
        latencies = [(c[0] - start, c[1]) for c in latencies]  # subtract start
        for y, (s, d) in enumerate(latencies):
            ax.plot([s, s + d], [y, y], c='red', linewidth=1)

        title = profile.title
        title = title.replace('%', r'\%').replace(' ', r'\ \ \ ')
        ax.set_title(title, loc='left', fontweight='normal', color='blue')

        exec_text = f'{profile.virtual_machine.exec / 1_000_000_000.0:>.2f} executing'
        compilation_text = f'{profile.compilation_service.total / 1_000_000_000.0:>.2f} compiling'
        waiting_text = f'{profile.virtual_machine.waiting / 1_000_000_000.0:>.2f} waiting'
        static_duration = (profile.virtual_machine.static_end - profile.virtual_machine.static_start) / 1_000_000_000.0
        static_text = f'{static_duration :>.2f} static'
        x = exec_end + (.005 * max_x)
        ax.text(x=x, y=.75 * max_y, s=exec_text, ha='left', va='bottom', color='green', fontsize='small')
        ax.text(x=x, y=.5 * max_y, s=compilation_text, ha='left', va='bottom', color='blue', fontsize='small')
        ax.text(x=x, y=.25 * max_y, s=waiting_text, ha='left', va='bottom', color='red', fontsize='small')
        if static_duration > 0:
            ax.text(x=x, y=0 * max_y, s=static_text, ha='left', va='bottom', color='purple', fontsize='small')

        ax.grid(False)
        ax.margins(0)

    # x-axis
    ax.set_xlim(0, 1.2 * max_x)

    def format_func(value, _):  # show seconds
        sec = value / 1_000_000_000
        if sec == int(sec):
            return f'{int(sec)}'
        return f'{sec:.1f}'

    ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))

    # y-axis
    ax.set_ylim(-0.05 * max_y, 1.01 * max_y)

    ax.set_xlabel('Time (s)')
    ax.set_ylabel('Functions')

    plt.tight_layout()

    plt.savefig(timelines_dir / f'{benchmark.mode}.pdf', format='pdf', dpi=320, bbox_inches='tight', pad_inches=0.1)

    # plt.close()

    plt.show()

# plot_timelines(benchmark, profiles)
```

```python
report_dir = Path('./out/report')
report_dir.mkdir(exist_ok=True, parents=True)

# from jetstream.float_mm.benchmark import float_mm

for benchmark in all_benchmarks:
    print(f'benchmark={benchmark.mode}')

    candidates, metrics = spec_profiler.profile(benchmark)
    print(f'#candidates={len(candidates)}')

    before_comps = comp_profiler.profile(benchmark)
    after_comps = comp_profiler.profile(benchmark, ['--specialize'])

    before_exec = call_profiler.profile(benchmark)
    after_exec = call_profiler.profile(benchmark, ['--specialize'])

    report, keep = optimize_specialization(metrics)

    keep = [{'idx': idx, 'name': candidates[idx]} for idx in keep]
    json.dump(keep, open('/tmp/specs.json', 'w'))

    df = pd.DataFrame([vars(v) for v in report.values()])
    df = df.sort_values(by='net_saving', ascending=False)
    df = df.reset_index(drop=True)
    df.to_excel(report_dir / f'{benchmark.mode}.xlsx', index=False)

    sc.compile(benchmark, ['-O3', '--ext=off'])
    sap.generate(benchmark)
    schemes = [
        ['--sync=aot'],
        ['--sync=jit'],
        ['--async=dynamic'],
        ['--static=surplus'],
        ['--async=dynamic', '--static=surplus'],
        ['--async=dynamic', '--static=surplus', '--specialize'],
        ['--static=all'],
    ]
    profiles = []
    for scheme in schemes:
        print(f'scheme={scheme}')
        profile = vm.run(benchmark, scheme)
        if scheme == ['--sync=jit']:
            functions = surplus_maximizer(profile)
            json.dump(functions, open('/tmp/surplus.json', 'w'))
        profiles.append(profile)
    plot_timelines(benchmark, profiles)
```

```python
report_dir = Path('./out/report')
report_dir.mkdir(exist_ok=True, parents=True)

e2e_waiting_dir = Path('./out/e2e-waiting')
e2e_waiting_dir.mkdir(exist_ok=True, parents=True)

from jetstream.float_mm.benchmark import float_mm

# all_benchmarks = [float_mm, float_mm]
for benchmark in all_benchmarks:
    print(f'benchmark={benchmark.mode}')

    candidates, metrics = spec_profiler.profile(benchmark)
    print(f'#candidates={len(candidates)}')

    before_comps = comp_profiler.profile(benchmark)
    after_comps = comp_profiler.profile(benchmark, ['--specialize'])

    before_exec = call_profiler.profile(benchmark)
    after_exec = call_profiler.profile(benchmark, ['--specialize'])

    report, keep = optimize_specialization(metrics)

    keep = [{'idx': idx, 'name': candidates[idx]} for idx in keep]
    json.dump(keep, open('/tmp/specs.json', 'w'))

    df = pd.DataFrame([vars(v) for v in report.values()])
    df = df.sort_values(by='net_saving', ascending=False)
    df = df.reset_index(drop=True)
    df.to_excel(report_dir / f'{benchmark.mode}.xlsx', index=False)

    sc.compile(benchmark)
    sap.generate(benchmark)
    schemes = [
        ['--sync=aot'],
        ['--sync=jit'],
        ['--async=dynamic'],
        ['--static=surplus'],
        ['--async=dynamic', '--static=surplus'],
        ['--async=dynamic', '--static=surplus', '--specialize'],
        ['--static=all'],
    ]

    profiles = []
    avg_waitings = []
    avg_e2es = []
    for scheme in schemes:
        print(f'scheme={scheme}')
        waiting_times = []
        e2e_times = []
        for _ in range(10):
            profile = vm.run(benchmark, scheme)
            static_loading = profile.virtual_machine.static_end - profile.virtual_machine.static_start
            waiting_times.append(profile.virtual_machine.waiting + static_loading)
            e2e_times.append(profile.virtual_machine.e2e)
        avg_waiting = sum(waiting_times) / len(waiting_times)
        avg_e2e = sum(e2e_times) / len(e2e_times)

        avg_waitings.append(avg_waiting)
        avg_e2es.append(avg_e2e)

        if scheme == ['--sync=jit']:
            all_functions = profile.virtual_machine.times
            functions = surplus_maximizer(profile)
            print(f'#static={len(functions)}')
            json.dump(functions, open('/tmp/surplus.json', 'w'))

        profiles.append(profile)

    plot_timelines(benchmark, profiles)

    # normalize waiting times and e2e times by first scheme
    avg_waitings = [w / avg_waitings[1] for w in avg_waitings]
    avg_e2es = [e / avg_e2es[1] for e in avg_e2es]

    num_specialized = len(keep)
    static_percent = int(100 * (len(functions) / len(all_functions)))
    static_title = f'static({static_percent}%)'
    spec_title = f'spec({num_specialized})'
    # spec_title = f'spec(N/A)'

    schemes_names = ['aot', 'jit', 'async', static_title, f'async+{static_title}', f'async+{static_title}+{spec_title}',
                     'full static']
    # make the names latex compatible
    schemes_names = [name.replace('%', r'\%') for name in schemes_names]
    fig, ax = plt.subplots(figsize=(5, 5), dpi=320)
    for i, scheme in enumerate(schemes):
        ax.scatter(avg_waitings[i], avg_e2es[i], label=schemes_names[i], alpha=0.5)

    ax.set_xlim(0, 2)
    ax.set_ylim(0, 2)

    ax.set_xlabel('Normalized Waiting Time')
    ax.set_ylabel('Normalized End-to-End Time')
    ax.legend(fontsize='small')
    plt.tight_layout()
    plt.savefig(e2e_waiting_dir / f'{benchmark.mode}.pdf', format='pdf', dpi=320, bbox_inches='tight', pad_inches=0.1)
    plt.close()
    # plt.show()

    print()
    print()
```
